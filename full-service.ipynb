****Cleaning the input_message by removing certain special characters (<, \ ,", /) and Replacing > with a space****

```python
import re
```

```python
def clean_text(input_text):
    pattern = r'[<\"/]'
    cleaned_text1 = re.sub(pattern, '', input_text)
    cleaned_text = cleaned_text1.replace('>',' ')
    return cleaned_text
```

****Creating Custom Datasets, Data Loaders and Loading the Tokenizer and the Model****

```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel # For tokenizing and using the GPT-2 model
from torch.utils.data import DataLoader, Dataset # # PyTorch utilities for handling datasets and data loading
```

```python
# Custom dataset class inheriting from PyTorch's Dataset
class CustomDataset(Dataset):
    def __init__(self, data, tokenizer, max_length):
        # Initialize the custom dataset with the provided data, tokenizer and max sequence length
        self.data = data  # The dataset
        self.tokenizer = tokenizer  # The tokenizer for converting text into input IDs
        self.max_length = max_length  # The maximum number of tokens

    def __len__(self):
        # Return the number of samples in the dataset
        return len(self.data)

    def __getitem__(self, index):
        # Retrieve the cleaned input message from the data at the specified index
        seq = self.data.iloc[index]['cleaned_input_message']

        # Retrieve the target message from the same index
        target = self.data.iloc[index]['CCP_Message']

        # Concatenate the input message and the target message
        input_seq = f"{seq} {target}"

        # Tokenize the concatenated sequence
        # return_tensors='pt' to return PyTorch tensors
        # max_length to truncate or pad sequences to a fixed length
        # truncation='True' to ensure sequences are not longer than `max_length`
        # padding='max_length' to pad shorter sequences to `max_length`
        input_ids = self.tokenizer.encode(
            input_seq,
            return_tensors='pt',
            max_length=self.max_length,
            truncation=True,
            padding='max_length'
        )

        # Return the tokenized sequence as a tensor
        return input_ids
```

```python
# Load the tokenizer and the distilgpt2 model
tokenizer = GPT2Tokenizer.from_pretrained('distilgpt2')
model = GPT2LMHeadModel.from_pretrained('distilgpt2')
```

```python
# Set the pad token to the end-of-sequence (EOS) token
tokenizer.pad_token = tokenizer.eos_token
```

```python
# Define the maximum number of tokens, batch size and number of epochs
MAX_LENGTH = 1000
BATCH_SIZE = 2
EPOCHS = 3
```

```python
# Creating Custom Datasets for Training and Validation
train_dataset = CustomDataset(train_data, tokenizer, MAX_LENGTH)
val_dataset = CustomDataset(val_data, tokenizer, MAX_LENGTH)
```

```python
print(type(train_dataset[0]))
```

```python
# Create a data loader for the training dataset
# Batches data according to BATCH_SIZE
# Shuffles the data to randomize the order of samples for better generalization during training
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)

# Create a data loader for the validation dataset
# Batches data according to BATCH_SIZE
# Does not shuffle the data to maintain consistent order during validation
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)
```

```python
# Check if a CUDA-enabled GPU is available, otherwise use the CPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Move the model to the appropriate device (GPU or CPU) for computation
model.to(device)
```

```python
# Create an optimizer for the model
# Using the AdamW optimizer
# Set the learning rate to 2e-5 (0.00002)
optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)
```

****Training the model****

```python
# Iterate through the specified number of epochs
for epoch in range(EPOCHS):
    # Set the model to training mode
    model.train()
    total_loss = 0  # Initialize the total loss for this epoch

    # Loop through each batch in the training data
    for batch in train_loader:
        # Move the batch to the specified device (CPU/GPU)
        input_ids = batch.to(device)

        # Create labels from the input data by making a clone
        labels = input_ids.clone()

        # Reset the gradients before backpropagation
        optimizer.zero_grad()

        # Forward pass through the model
        outputs = model(input_ids, labels=labels)
        # Retrieve the loss from the outputs
        loss = outputs.loss
        
        # Backward pass for gradient computation
        loss.backward()
        
        # Update model parameters with the optimizer
        optimizer.step()

        # Accumulate the loss for this batch
        total_loss += loss.item()

    # Calculate the average loss for this epoch
    avg_train_loss = total_loss / len(train_loader)
    # Display the average training loss for this epoch
    print(f'Epoch {epoch + 1}, Average Training Loss: {avg_train_loss}')
```

****Evaluating the pre-trained model****

```python
# Set the model to evaluation mode
model.eval()

# Initialize total loss for validation
total_loss = 0

# Disable gradient calculation during validation
with torch.no_grad():
    # Loop through each batch in the validation data
    for batch in val_loader:
        # Move the batch to the specified device (CPU/GPU)
        input_ids = batch.to(device)

        # Create labels from the input data by making a clone
        labels = input_ids.clone()

        # Forward pass through the model for validation
        outputs = model(input_ids, labels=labels)
        
        # Retrieve the loss from the outputs
        loss = outputs.loss
        
        # Accumulate the loss for this batch
        total_loss += loss.item()

# Calculate the average loss for the validation set
avg_val_loss = total_loss / len(val_loader)

# Display the average validation loss
print(f'Average Validation Loss: {avg_val_loss}')
```

```python
# Function to generate a sequence using the pre-trained GPT-2 model
def predict(seq):
    # Set the model to evaluation mode 
    model.eval()
    
    # Encode the input sequence into token IDs, returning PyTorch tensors, and move to the correct device
    input_ids = tokenizer.encode(seq, return_tensors='pt').to(device)
    
    # Generate output from the encoded input
    output = model.generate(
        input_ids,
        max_length=1000,  
        num_return_sequences=1,  
        pad_token_id=tokenizer.eos_token_id  
    )
    
    # Decode the generated output to convert token IDs back to string
    decoded_output = tokenizer.decode(output[0], skip_special_tokens=True)
    
    # Remove the original input sequence from the generated output
    output_sequence = decoded_output[len(seq):].strip()
    
    # Return the final output sequence
    return output_sequence
```

****Calculating the Accuracy****

```python
# Initialize a variable to keep track of the number of correct predictions
som = 0

# Loop through the samples in the validation data
for i in range(100):
    # Get the original CCP message and the predicted message
    original_message = val_data.iloc[i]['CCP_Message']
    predicted_message = predict(val_data.iloc[i]['cleaned_input_message'])
    
    # Check if the predicted message matches the original message
    if original_message == predicted_message:
        # If they match, increment the counter for correct predictions
        som += 1

# Calculate the accuracy as the proportion of correct predictions
accuracy = som / 100

# Print the accuracy
print('Accuracy:', accuracy)
```

****Identifying Causes of Mismatched Predictions****

```python
# Loop through the samples in the validation data
for i in range(100):
    # Predict the message based on the 'cleaned_input_message'
    predicted_message = predict(val_data.iloc[i]['cleaned_input_message'])
    
    # Get the corresponding original 'CCP_Message' from the dataset
    original_message = val_data.iloc[i]['CCP_Message']
    
    # Check if the predicted message does not match the original
    if predicted_message != original_message:
        print('Message', i)  # Identify the index of the mismatched message
        print('Predicted message:')  
        print(predicted_message)  # Print the predicted message
        print('Test message:')  
        print(original_message)  # Print the original test message
```
